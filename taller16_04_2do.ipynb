```python
import numpy as np
from scipy.optimize import fsolve

# a) clasificacion de variables
print("a) VARIABLES:")
print("Variable independiente: n (semana)")
print("Variables dependientes: S_n (sillas), M_n (mesas)")
print()

# b) clasificacion del problema
print("b) CLASIFICACION:")
print("Modelo lineal y discreto")
print()

# c) desarrollo del problema
#se definen las variables de sillas y mesas iniciales y en que semana seria
def evolucion(S0, M0, semanas):
    S = [S0]
    M = [M0]
#se repite el proceso por varias semanas
    for i in range(semanas):
        S_next = 0.6*S[i] + 0.2*M[i] + 40
        M_next = 0.1*S[i] + 0.5*M[i] + 20
#los valores se van añadiendo a la lista
        S.append(S_next)
        M.append(M_next)

    return S, M
#establece los valores iniciales en la primera semana y calcula hasta la 2
S0, M0 = 200, 80
S, M = evolucion(S0, M0, 2)

print("c) RESULTADOS:")
print(f"Semana 0: S0 = {S[0]:.2f}, M0 = {M[0]:.2f}")
print(f"Semana 1: S1 = {S[1]:.2f}, M1 = {M[1]:.2f}")
print(f"Semana 2: S2 = {S[2]:.2f}, M2 = {M[2]:.2f}")
print()

# d) equilibrio

def sistema(vars):
    S, M = vars
    eq1 = S - (0.6*S + 0.2*M + 40)
    eq2 = M - (0.1*S + 0.5*M + 20)
    return [eq1, eq2]

# valor inicial para buscar solucion
sol = fsolve(sistema, [100, 100])

print("d) EQUILIBRIO:")
print(f"S = {sol[0]:.2f}")
print(f"M = {sol[1]:.2f}")
print()
```
