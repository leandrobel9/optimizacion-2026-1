```python
import numpy as np
from scipy.integrate import solve_ivp
# a) clasificacion variables, y parametros

print("a) VARIABLES Y PARAMETROS:")
print("Variable independiente: t (tiempo en dias)")
print("Variable dependiente: I(t) (infectados)")
print("Parametros: beta = 0.00003, gamma = 0.1, N = 10000")
print()

# b) clasificacion del problema

print("b) CLASIFICACION:")
print("Modelo no lineal (termino I^2)")
print()

# c) condicion inicial
I0 = 50
print("c) CONDICION INICIAL:")
print(f"I(0) = {I0}")
print()

#EDO
beta = 0.00003
gamma = 0.1
N = 10000

def modelo(t, I):
    return beta*(N - I)*I - gamma*I

# solucion
t_span = (0, 100)  # 100 dias
t_eval = np.linspace(0, 100, 200)

sol = solve_ivp(modelo, t_span, [I0], t_eval=t_eval)

#
# d) equilibrio
#
I_eq1 = 0
I_eq2 = N - gamma/beta

print("d) EQUILIBRIOS:")
print(f"I = {I_eq1}")
print(f"I = {I_eq2:.2f}")
print()

# resultados

print("EVOLUCION (algunos valores):")
print(f"I(0) = {sol.y[0][0]:.2f}")
print(f"I(50) = {sol.y[0][100]:.2f}")
print(f"I(100) = {sol.y[0][-1]:.2f}")
print()
```
