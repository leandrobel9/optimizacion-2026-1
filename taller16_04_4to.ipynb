```python
import numpy as np
from scipy.optimize import fsolve


# a) clasificacion de variables, y parametros

print("a) VARIABLES Y PARAMETROS:")
print("Variable independiente: n (años)")
print("Variable dependiente: P_n (poblacion de peces)")
print("Parametros: r = 0.4, K = 1000, H = 80")
print()


# b) clasificacion del problema

print("b) CLASIFICACION:")
print("Modelo no lineal (termino P^2)")
print("Modelo discreto")
print()

#
# c) calculo DE P1 Y P2

r = 0.4
K = 1000
H = 80

def modelo(P):
    return P + r*P*(1 - P/K) - H

P0 = 400

P1 = modelo(P0)
P2 = modelo(P1)

print("c) RESULTADOS:")
print(f"P0 = {P0:.2f}")
print(f"P1 = {P1:.2f}")
print(f"P2 = {P2:.2f}")
print()

# d) equilibrio (P_{n+1} = P_n)

def equilibrio(P):
    return modelo(P) - P

sol = fsolve(equilibrio, [200, 800])  # dos aproximaciones

print("d) EQUILIBRIOS:")
print(f"P ≈ {sol[0]:.2f}")
print(f"P ≈ {sol[1]:.2f}")
print()

# e) interpretacion

print("e) INTERPRETACION:")
print("El equilibrio menor es inestable (riesgo de colapso)")
print("El equilibrio mayor es estable (poblacion sostenible)")
print()
```
