```python
import pulp

# a) variables de decision

print("a) VARIABLES DE DECISION:")
print("x1: toneladas de fertilizante F1")
print("x2: toneladas de fertilizante F2")
print()

#se crea el objeto que vamos a maximimizar

modelo = pulp.LpProblem("Mezcla_Fertilizantes", pulp.LpMaximize)


# se crean las variables y se les pone no negatividad

x1 = pulp.LpVariable('F1', lowBound=0)
x2 = pulp.LpVariable('F2', lowBound=0)


# b) funcion objetivo

modelo += 180*x1 + 220*x2

print("b) FUNCION OBJETIVO:")
print("Maximizar Z = 180x1 + 220x2")
print()

# c) restricciones

modelo += 3*x1 + 4*x2 <= 120
modelo += 2*x1 + x2 <= 80
modelo += 5*x1 + 3*x2 <= 150

print("c) RESTRICCIONES:")
print("3x1 + 4x2 <= 120  (Compuesto A)")
print("2x1 + x2 <= 80    (Compuesto B)")
print("5x1 + 3x2 <= 150  (Compuesto C)")
print("x1 >= 0, x2 >= 0  (No negatividad)")
print()

# d) clasificacion
print("d) CLASIFICACION DEL MODELO:")
print("Modelo de programacion lineal")
print("Tipo continuo")
print()

# resolver

modelo.solve()

# muestra resultados

print("="*45)
print("RESULTADO OPTIMO")
print("="*45)

print(f"Estado del modelo: {pulp.LpStatus[modelo.status]}")
print()

print(f"Produccion optima de F1 (x1): {pulp.value(x1):.2f} toneladas")
print(f"Produccion optima de F2 (x2): {pulp.value(x2):.2f} toneladas")
print()
#muestra el valor optimo de la funcion objetivo
print(f"Ganancia maxima: ${pulp.value(modelo.objective):.2f}")
print("="*45)
```
